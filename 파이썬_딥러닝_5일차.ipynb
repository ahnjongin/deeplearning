#IMDB RNN

```python
#모듈 임포트
import keras
from keras import models, layers
from keras import backend
from keras import datasets
from keras.datasets import imdb
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing import sequence
```

```python
#데이터 로딩
(x_train, y_train), (x_test, y_test)= imdb.load_data(num_words= 20000)
```

```python
#단어와 단어에 부여한 인덱스 테이블 로딩
word_to_index=imdb.get_word_index()
```

```python
index_to_word={}
for key, value in word_to_index.items():
  index_to_word[value]=key
  print(key + "," + str(value))
```

```python
#독립변수로 사용할 수 있게 댓글에 사용한 단어 중에 80개만 사용하게 정제화(부족하면 0으로 Padding)
x_train2= sequence.pad_sequences(x_train, maxlen=80)
x_test2 = sequence.pad_sequences(x_test, maxlen=80)
```

```python
#입력 게층 만들기
x = layers.Input(80,)
```

```python
#은닉 계층(임베디드 계층 만들기)
h= layers.Embedding(20000,128)(x)
```

```python
#lstm 계층
lstm = layers.LSTM(128, dropout=0.2, recurrent_dropout=0.2)
h=lstm(h)
```

```python
#출력 계층
dense = layers.Dense(1, activation='sigmoid')
y= dense(h)
```

```python
#모델 생성
model= models.Model(x,y)
```

```python
#학습 가능하게 모델 컴파일
model.compile(loss= 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#계층확인
model.summary()
```

```python
#학습
model.fit(x_train2, y_train, epochs=3, batch_size=32)
```

```python
score = model.evaluate(x_test2, y_test, batch_size=32)
score
```
